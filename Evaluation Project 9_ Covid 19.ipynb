### Column Description:-

1. Province_State - The name of the State within the USA.
2. Country_Region - The name of the Country (US).
3. Last_Update - The most recent date the file was pushed.
4. Lat - Latitude.
5. Long_ - Longitude.
6. Confirmed - Aggregated confirmed case count for the state.
7. Deaths - Aggregated Death case count for the state.
8. Recovered - Aggregated Recovered case count for the state.
9. Active - Aggregated confirmed cases that have not been resolved (Active = Confirmed - Recovered - Deaths).
10. FIPS - Federal Information Processing Standards code that uniquely identifies counties within the USA.
11. Incident_Rate - confirmed cases per 100,000 persons.
12. People_Tested - Total number of people who have been tested.
13. People_Hospitalized - Total number of people hospitalized.
14. Mortality_Rate - Number recorded deaths * 100/ Number confirmed cases.
15. UID - Unique Identifier for each row entry.
16. ISO3 - Officialy assigned country code identifiers.
17. Testing_Rate - Total number of people tested per 100,000 persons.
18. Hospitalization_Rate - Total number of people hospitalized * 100/ Number of confirmed cases.

We can see outliers in all columns except Province State. Hence we have to treat them

```python
#removing the outliers
from scipy.stats import zscore
z_score=abs(zscore(ds))
print(ds.shape)

ds_final=ds.loc[(z_score<3).all(axis=1)]
print(ds_final.shape)
```

```python
ds=ds_final
ds.shape
```

```python
#lets chcek for skewness
print(ds.skew())
#lets visualize graphically
for i in ds.columns:
    plt.figure()
    sns.distplot(ds[i])
    plt.show()
```

```python
#we can see skewness in most of the columns hence we have to treat them.
for col in ds.columns:
    if ds[col].skew()>0.55:
        ds[col]=np.cbrt(ds[col])
```

```python
#checking skewness again
ds.skew()
```

Skewness has been treated.

### Training the model

```python
y=ds['Deaths']
y
```

```python
ds_x=ds.drop(columns='Deaths',axis=1)
ds_x
```

```python
#we have to standardize the input variables
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(ds_x)
x=pd.DataFrame(x,columns=ds_x.columns)
```

```python
x
```

```python
#importing libraries for model creation
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=42)
print(x_train.shape,'\t',x_test.shape)
print(y_train.shape,'\t',y_test.shape)
```

```python
#creating a for loop for all the models
model=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),SVR(),KNeighborsRegressor(),DecisionTreeRegressor()]
for m in model:
    m.fit(x_train,y_train)
    print('score of ',m,'is:',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print('Metrics:')
    print('mean absolute error:',mean_absolute_error(y_test,predm))
    print('mean squared error:',mean_squared_error(y_test,predm))
    print('r2 score:',r2_score(y_test,predm))
    print('END OF THE MODEL----------------------------------- ')
    print('\n')
```

## Ensemble Methods

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
#creating a for loop for all the models
model=[RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
for m in model:
    m.fit(x_train,y_train)
    print('score of ',m,'is:',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print('Metrics:')
    print('mean absolute error:',mean_absolute_error(y_test,predm))
    print('mean squared error:',mean_squared_error(y_test,predm))
    print('r2 score:',r2_score(y_test,predm))
    print('END OF THE MODEL----------------------------------- ')
    print('\n')
```

Of ths above models Ridge Regressor has the best r2 sccore and least errors. lets do parameter tuning to improve the scores.

```python
#tuning using gridsearch cv
ridge= Ridge()
from sklearn.model_selection import GridSearchCV
params_Ridge = {'alpha': [1,0.1,0.01,0.001,0.0001] , 'random_state': range(42,100)}
Ridge_GS = GridSearchCV(ridge, param_grid=params_Ridge, scoring='r2')
Ridge_GS.fit(x_train,y_train)
Ridge_GS.best_params_
```

```python
#ridge model after parameter tuning
ridge=Ridge(alpha=0.1,random_state=42)
ridge.fit(x_train,y_train)
print('score of ',ridge,'is:',ridge.score(x_train,y_train))
predrd=ridge.predict(x_test)
print('Metrics:')
print('mean absolute error:',mean_absolute_error(y_test,predrd))
print('mean squared error:',mean_squared_error(y_test,predrd))
print('r2 score:',r2_score(y_test,predrd))
print('END OF THE MODEL----------------------------------- ')
print('\n')
```

```python
#cross validate ridge
from sklearn.model_selection import cross_val_score
score=cross_val_score(ridge,x,y,cv=4,scoring='r2')
print('Score :',score)
print('Mean Score :',score.mean())
print('Standard Deviation :',score.std())
```

The model is having a mean score of 0.95 .Hence the model is not over-fitted and under-fitted. The Ridge Regressor model can be considered as the best model for this Covid19 dataset.
